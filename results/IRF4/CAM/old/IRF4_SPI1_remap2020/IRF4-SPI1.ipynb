```python
import pandas as pd
```

```python
wt = pd.read_csv("./WT/outputs.tsv", sep="\t", header=0)
wt["Patient"] = "WT"
wt
```

```python
t95r = pd.read_csv("./T95R/outputs.tsv", sep="\t", header=0)
t95r["Patient"] = "T95R"
t95r
```

```python
df = pd.concat([wt, t95r])
df
```

```python
import seaborn as sns
```

```python
sns.violinplot(data=df, x="Patient", y="IRF4")
```

```python
from scipy import stats
```

```python
stats.ttest_ind(df[df.Patient == "WT"]["IRF4"], df[df.Patient == "T95R"]["IRF4"])
```

```python
print(df[df.Patient == "WT"]["IRF4"].mean())
print(df[df.Patient == "T95R"]["IRF4"].mean())
```
